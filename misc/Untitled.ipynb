```python
import networkx as nx
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('/home/laura/Documents/CODE/graph_genetics/data/graph_datasets/PETandDX/AD_PPI_snap_brain_missense.pkl', 'rb') as f:
    graphs = pickle.load(f)

for g in graphs:
    G = nx.Graph(g)
    print(G.graph)
#     print(G.nodes(data=True))
    print(G.graph)
    
    nodes_remove = []
    for node in G.nodes(data=True):
        attr = int(node[1]['node_attr'])
        if attr == 0:
            nodes_remove.append(node[0])
#             print(node)
    G.remove_nodes_from(nodes_remove)

    plt.figure(figsize=(12, 8))
    nx.draw(G,
            with_labels=True,
            node_size=800,
            font_color='white')

    plt.show()
```
