```python
import os, sys, datetime, pickle
import pandas as pd
import networkx as nx
from networks import randomize_network
from networks.utils import pvalues_random
from create_datasets import create_nx_datasets
```

**1. Create random networks**

Create 100 random networks for each randomization method ("Shuffled" and "Rewired") and for each gene set (AD and NDD).

```python
infile = f'data/other_networks/AD_SNAP_PPI_brain.edgelist'
original = rn.load_graph(infile)

for i in range(1, 101):
    shuffled = randomize_network.shuffle_nodes(original)
    nx.write_edgelist(shuffled, f'data/random_networks/AD_PPI_shuffled{i}_edgelist.txt')

    rewired = randomize_network.generate_RDPN(infile)
    nx.write_edgelist(shuffled, f'data/random_networks/AD_PPI_rewired{i}_edgelist.txt')
```

**2. Create graph datasets**

Create the correspoding graph-datasets for each random network. This part takes time because the code builds 400 different datasets (2 randomisation methods x 100 random networks x 2 different targets)

```python
dataset = 'ADNI'
targets = ['PET', 'PETandDX']
networks = ['shuffled', 'rewired']

for target in targets:
    for network in networks:
        for i in range(1, 101):

            outdir = f'data/graph_datasets/{target}/random'

            start_time = datetime.datetime.now()
            print()

            result_nodes = create_nx_datasets.main('data', dataset, target, 'AD', network, 'missense', i)
            print('Coding: number of missense variants per node')

            outfile = f'{outdir}/AD_PPI_{network}{i}_missense.pkl'
            print('Resulting dataset saved at:', outfile)
            print()

            with open(outfile, 'wb') as f:
                pickle.dump(result_nodes, f)

            result_nodes_time = datetime.datetime.now()
            print('Processing time:', result_nodes_time - start_time)
            print('\n\n')
```

**3. Graph classification with GNNs**

We then evaluated the best GNNs for each target in the framework called [GraphGym](https://github.com/snap-stanford/GraphGym) (You *et al.*, 2020).

Configuration and grid files employed are in the subdirectory [graphgym_files](graphgym_files).

Summarized results obtained by GraphGym and other models are in [results/GNN_comparison](results/GNN_comparison)

**4. Statistical analysis original performance *vs.* random performances**

We computed p-values with a 1-sample t-test comparing in each fold the original AUC value obtained against 100 random values (100 random datasets) of each randomization method for each target. 

- Results for PET target with datasets using [Shuffled](results/2022_04_networks_PET_random_shuffled.csv) and [Rewired](results/2022_04_networks_PET_random_rewired.csv) methods.
- Results for PET&DX target with datasets using [Shuffled](results/2022_04_networks_PETandDX_random_shuffled.csv) and [Rewired](results/2022_04_networks_PETandDX_random_rewired.csv) methods.

```python
pet_results          = pd.read_csv('results/2022_03_networks_PET.csv', index_col='fold')
pet_shuffled_results = pd.read_csv('results/2022_04_networks_PET_random_shuffled.csv', index_col='fold')
pet_rewired_results  = pd.read_csv('results/2022_04_networks_PET_random_rewired.csv', index_col='fold')

pvals_shuffled = pvalues_random(pet_results, pet_shuffled_results, 'shuffled')
pvals_shuffled = pvalues_random(pet_results, pet_rewired_results, 'rewired')
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline

def jaccard_similarity(g, h):
    'Compute Jaccard similarity index using the ineractions present in each of the two compared networks.'
    i = set(g).intersection(h)    
    j = len(i) / (len(g) + len(h) - len(i))
    return j

G_original = nx.read_edgelist('data/other_networks/AD_SNAP_PPI_brain.edgelist')

shuffled_graphs = []
rewired_graphs  = []

for i in range(1, 101):
    G_shuffled = nx.read_edgelist(f'data/random_networks/AD_PPI_shuffled{i}_edgelist.txt')
    shuffled_graphs.append(G_shuffled)

for i in range(1, 101):
    G_rewired = nx.read_edgelist(f'data/random_networks/AD_PPI_rewired{i}_edgelist.txt')
    rewired_graphs.append(G_rewired)

# print(len(shuffled_graphs))
# print(len(rewired_graphs))

results_shuffled = {}
results_rewired = {}

i = 1
for G_shuffled in shuffled_graphs:
    jaccard_shuffled = jaccard_similarity(G_original.edges(), G_shuffled.edges())
    results_shuffled[f'AD_PPI_shuffled{i}_missense'] = jaccard_shuffled
    i += 1

i = 1
for G_rewired in rewired_graphs:
    jaccard_rewired = jaccard_similarity(G_original.edges(), G_rewired.edges())
    results_rewired[f'AD_PPI_rewired{i}_missense'] = jaccard_rewired
    i += 1
    
print()

# results_shuffled
```

```python
pet_rewired_results['sim'] = pet_rewired_results['dataset'].map(results_rewired)
pet_shuffled_results['sim'] = pet_shuffled_results['dataset'].map(results_shuffled)

plt.figure(figsize=(10,10))

for i in range(10):
    i += 1
    
    aucs = pet_shuffled_results[pet_shuffled_results.index==i]['auc'].tolist()
    sims = pet_shuffled_results[pet_shuffled_results.index==i]['sim'].tolist()
    sns.regplot(x=sims, y=aucs, ci=None, label=f'Fold {i}')
    
plt.title(f'Shuffled - Jaccard similarity vs. AUC performance')
plt.legend()
plt.show()

plt.figure(figsize=(10,10))
for i in range(10):
    i += 1
    
    aucs = pet_rewired_results[pet_rewired_results.index==i]['auc'].tolist()
    sims = pet_rewired_results[pet_rewired_results.index==i]['sim'].tolist()
    sns.regplot(x=sims, y=aucs, ci=None, label=f'Fold {i}')
    
plt.title(f'Rewired - Jaccard similarity vs. AUC performance')
plt.ylabel('AUC')
plt.xlabel('Jaccard index')
plt.legend()
plt.show()
```
