```python
import os, sys, datetime, pickle
import pandas as pd
import networkx as nx
from networks import randomize_network
from networks.networks_stats import pvalues_random
from create_datasets import create_nx_datasets
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Create random networks

Create 100 random networks for each randomization method ("Shuffled" and "Rewired") using AD PPT-Ohmnet network.

```python
infile = f'data/other_networks/AD_SNAP_PPI_brain.edgelist'
original = rn.load_graph(infile)

for i in range(1, 101):
    shuffled = randomize_network.shuffle_nodes(original)
    nx.write_edgelist(shuffled, f'data/random_networks/AD_PPI_shuffled{i}_edgelist.txt')

    rewired = randomize_network.generate_RDPN(infile)
    nx.write_edgelist(shuffled, f'data/random_networks/AD_PPI_rewired{i}_edgelist.txt')
```

### 2. Create graph datasets

Create the corresponding graph-datasets for each random network. This part takes time because the code builds 400 different datasets (2 randomisation methods x 100 random networks x 2 different labels in ADNI)

```python
dataset = 'ADNI'
targets = ['PET', 'PETandDX']
networks = ['shuffled', 'rewired']

for target in targets:
    for network in networks:
        for i in range(1, 101):

            outdir = f'data/graph_datasets/{target}/random'

            start_time = datetime.datetime.now()
            print()

            result_nodes = create_nx_datasets.main('data', dataset, target, 'AD', network, 'missense', i)
            print('Coding: number of missense variants per node')

            outfile = f'{outdir}/AD_PPI_{network}{i}_missense.pkl'
            print('Resulting dataset saved at:', outfile)
            print()

            with open(outfile, 'wb') as f:
                pickle.dump(result_nodes, f)

            result_nodes_time = datetime.datetime.now()
            print('Processing time:', result_nodes_time - start_time)
            print('\n\n')
```

### 3. Graph classification with GNNs

We trained and tested the best GNNs obtained for each label as done with the original graph datasets.

Configuration and grid files employed are in the subdirectory [graphgym_files](graphgym_files). The GNNs configuration in this case was the same as the best configuration found with AD PPT-Ohment graph datasets.

We run GraphGym 10 times, with 10 different splits generated by 10-Fold Stratified Cross-Validation.

### 4. Statistical analysis original performance *vs.* random performances

We computed p-values with a 1-sample t-test comparing in each fold the original AUC value obtained against 100 random values (100 random datasets) of each randomization method for each target. 

```python
pet_results          = pd.read_csv('results/2022_03_networks_PET.csv', index_col='fold')
pet_shuffled_results = pd.read_csv('results/2022_04_networks_PET_random_shuffled.csv', index_col='fold')
pet_rewired_results  = pd.read_csv('results/2022_04_networks_PET_random_rewired.csv', index_col='fold')

pvals_shuffled = pvalues_random(pet_results, pet_shuffled_results, 'shuffled')
pvals_shuffled = pvalues_random(pet_results, pet_rewired_results, 'rewired')
```

```python
petdx_results          = pd.read_csv('results/2022_03_networks_PETandDX.csv', index_col='fold')
petdx_shuffled_results = pd.read_csv('results/2022_04_networks_PETandDX_random_shuffled.csv', index_col='fold')
petdx_rewired_results  = pd.read_csv('results/2022_04_networks_PETandDX_random_rewired.csv', index_col='fold')

pvals_shuffled = pvalues_random(petdx_results, petdx_shuffled_results, 'shuffled')
pvals_shuffled = pvalues_random(petdx_results, petdx_rewired_results, 'rewired')
```

```python
pet = pet_results.loc[pet_results['dataset'] == 'AD PPT-Ohmnet']
petdx = petdx_results.loc[petdx_results['dataset'] == 'AD PPT-Ohmnet']

pet_rewired_results['dataset'] = 'AD PPT-Ohmnet Rewired'
pet_shuffled_results['dataset'] = 'AD PPT-Ohmnet Shuffled'

petdx_rewired_results['dataset'] = 'AD PPT-Ohmnet Rewired'
petdx_shuffled_results['dataset'] = 'AD PPT-Ohmnet Shuffled'
```

```python
pet_all   = pd.concat([pet, pet_shuffled_results, pet_rewired_results])
petdx_all = pd.concat([petdx, petdx_shuffled_results, petdx_rewired_results])

pet_all
```

```python
pet_all.reset_index(inplace=True)
petdx_all.reset_index(inplace=True)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(18, 8))

sns.boxplot(x = 'fold',
            y = 'auc',
            hue = 'dataset',
            data = pet_all,
            ax = ax1,
            palette = sns.color_palette(["#F564E3", "#64f5bf", "#9bf564"]),
            showmeans = True, meanprops = {"markerfacecolor":"white", "markeredgecolor":"black", "markersize":"10"})


sns.boxplot(x = 'fold',
            y = 'auc',
            hue = 'dataset',
            data = petdx_all,
            ax = ax2,
            palette = sns.color_palette(["#F564E3", "#64f5bf", "#9bf564"]),
            showmeans = True, meanprops = {"markerfacecolor":"white", "markeredgecolor":"black", "markersize":"10"})

ax1.set_ylim(0.5, 1.0)
ax2.set_ylim(0.5, 1.0)

ax1.set_ylabel('AUC', fontsize=16)
ax2.set_ylabel('AUC', fontsize=16)

ax1.set_xlabel('Fold number', fontsize=16)
ax2.set_xlabel('Fold number', fontsize=16)

ax1.tick_params(labelsize=16)
ax2.tick_params(labelsize=16)

ax1.set_title('(a) PET label - ADNI dataset', fontsize=16)
ax2.set_title('(b) PET&DX label - ADNI dataset', fontsize=16)

plt.tight_layout()

# plt.savefig(f'figures/figure4.png', dpi=500)
plt.savefig(f'figures/figure4.eps', dpi=500)
```
